# 掌控板与pinpong库的实践准备

mPython掌控是一块MicroPython微控制器板。专为物联网设计，板载ESP-WROOM-32双核芯片，支持WiFi和蓝牙双模通信。



### 安装驱动

安装软件时，首先观察掌控板版本，如果使用的是2.0以下版本，需要安装CP210x的驱动，可手动下载并安装（for Windows、MacOS），链接如下。

https://www.silabs.com/documents/public/software/CP210x_Windows_Drivers.zip

https://www.silabs.com/documents/public/software/Mac_OSX_VCP_Driver.zip
x64对应Windows 64位，x86对应Windows 32位

如果使用的掌控板是2.0及以上版本，掌控板的使用的是9102的驱动，可手动下载并安装（for Windows、MacOS），链接如下。
Windows版本链接：https://pan.baidu.com/s/1IcMIWUVdaTU1l17IF7BYmw 提取码：pong 
Mac版本链接：链接：https://pan.baidu.com/s/1Q7L7w-m6pzst9vr6jAm_Qw 提取码：pong 

### 安装pinpong库

pinpong库是一套控制开源硬件主控板的Python库，借助于pinpong库，直接用Python代码就能给各种常见的开源硬件编程。
pinpong库的名称由“Pin”和“Pong”组成，“Pin”指引脚，“PinPong”为“乒乓球”的谐音，指信号的往复。

pinpong库依赖Python，因此请确保电脑安装了Python3。

```python
!pip install pinpong
```

如果网络较慢安装不成功，可以指定国内源进行安装。

```python
!pip install pinpong -i http://mirrors.aliyun.com/pypi/simple/
```
